```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
solar_db = pd.read_excel('/content/solar_db.xlsx')

solar_db.head()
```

```python
solar_db.shape
```

```python
solar_db.isnull().sum()
```

```python
solar_db.describe()
```

```python
solar_db['Solar Production [MWh]'].mean()
```

```python
solar_db['Solar Production [MWh]'].median()
```

```python
solar_db['Solar Production [MWh]'].mode()
```

```python
# Tamanho da imagem
plt.figure( figsize=(25, 8) )

# Style
plt.style.use('ggplot')

# Grafico 2
plt.title( 'Solar Daily Energy Producing Hourly', loc='center', fontsize=13 )
sns.lineplot( data=solar_db, x='Date', y='Solar Production [MWh]',ci=None, lw=7, color='#8A2BE2', alpha=0.70);

# Labels
plt.xlabel('Date');
plt.ylabel('Solar Production [MWh]');
```

```python
# Tamanho da imagem
plt.figure( figsize=(25, 8) )

# Style
plt.style.use('ggplot')

# Grafico 2
plt.title( 'Solar Daily Energy Producing Hourly', loc='center', fontsize=13 )
sns.scatterplot( data=solar_db, x='Date', y='Solar Production [MWh]',ci=None, lw=7, color='#8A2BE2', alpha=0.70);

# Labels
plt.xlabel('Date');
plt.ylabel('Solar Production [MWh]');
```

```python
# Tamanho da imagem
plt.figure( figsize=(25, 8) )

# Style
plt.style.use('ggplot')

# Grafico 2
plt.title( 'Solar Daily Energy Producing Hourly', loc='center', fontsize=13 )
sns.histplot( data=solar_db, x='Date', y='Solar Production [MWh]', lw=7, color='#8A2BE2', alpha=0.70);

# Labels
plt.xlabel('Date');
plt.ylabel('Solar Production [MWh]');
```
